# **3. Find the percentage of men and women who survived. Plot them in a pie cahrt.**

## Finding the Percentage

```python
df_copy = df.copy()
survived = df_copy[df_copy['Survived'] == 1]

male = survived[survived['Sex'] == 'male']
female = survived[survived['Sex'] == 'female']

male_prcnt = (male.shape[0] / survived.shape[0]) * 100
female_prcnt = (female.shape[0] / survived.shape[0]) * 100

prcnt_dict = {
    "Gender" : ["Male", "Female"], 
    "Percentage" : [male_prcnt, female_prcnt]
}

prcnt_df = pd.DataFrame(prcnt_dict)
prcnt_df
```

## Ploting the Dataframe

```python
plt.style.use('fivethirtyeight')
explode = [0.1, 0.2]

plt.pie(prcnt_df['Percentage'], labels = prcnt_df['Gender'], explode=explode, shadow=True, startangle=90, autopct='%1.1lf%%')
```

# **7. Find the number of passangers who embarked from each destination and show them in an hbar graph.**

```python
embark = df.copy()
embark['Count'] = embark.index

embark_grp = embark.groupby('Embarked')['Count'].count()
embark_grp_df= pd.DataFrame(embark_grp).reset_index()
embark_grp_df.drop(1, axis=0, inplace = True)
embark_grp_df
```

```python
plt.barh(embark_grp_df['Embarked'], embark_grp_df['Count'])
plt.title("Embark graph of passengers")
plt.xlabel("Number of Passengers")
plt.ylabel("Embarked Place")
```
