# **2. (b) Handle the missing data in whatever manner that makes sense to you.**

## Handleing NaN of 'Age' column

```python
meanAge_ = df.copy().groupby(['Pclass', 'Sex'])['Age'].count()
meanAge_
```

1. Handled the missing values with mean values of the groups accourding to Pclass and gender. 
2. Above code shows there are 6 gropus. We calculated the means of each groups and replaced the NaNs of those groups with them.

```python
meanAge = df.groupby(['Pclass', 'Sex'])['Age'].transform('mean')
df['Age'] = df['Age'].fillna(meanAge)
df
```

Checking if any NaN left

```python
df[df['Age'].isnull() == True]
```

## Handleing NaN of 'Fare' column

Group-Specific Fill: If you want to fill NaN values within groups (e.g., based on a specific column value), you can use groupby() along with fillna()

```python
df['Fare'].unique().shape
```

```python
df_copy = df.copy()
df_copy["Count"] = df_copy.index
fare_group = df_copy.groupby([ 'Fare'])['Count'].count()
fare_group = pd.DataFrame(fare_group)
fare_group
```

```python
df['Fare'].dtype
```

```python
fare_group.shape
```

1. Handled the missing values with mean values of the groups accourding to Pclass, SibSp and Parch. 
2. Above code shows there are 38 gropus. We see that this groups has common fairs, so we replaced the NaNs of those groups with thst common fare.

```python
df['Fare'] = df.groupby(['Pclass', 'SibSp', 'Parch'])['Fare'].fillna(method='ffill')
df
```
